```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
month = "02"
year = "2024"
```

```python
outcome_data = pd.read_excel(rf"C:\Users\79165\Downloads\outcome_{month}.{year}.xlsx")
outcome_data
```

```python
outcome_data["День"]=[int(x.split()[0]) for x in outcome_data["Дата"]]
```

```python
fig, ax = plt.subplots(constrained_layout=True)
sns.lineplot(
    data = outcome_data,
    x = "День",
    y = "Сумма",
    hue = "Категория",
    ax=ax
)
ax.legend()
plt.show()
    
```

```python
fig, ax = plt.subplots(constrained_layout=True)

sns.barplot(
    data=outcome_data,
    y = "Категория",
    x = "Сумма",
    orient="h",
    estimator = "sum",
    errorbar=None,
    ax=ax
)

plt.show()
```
